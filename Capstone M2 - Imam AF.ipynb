## **Overview Transjakarta** 

Transjakarta (terkadang ditulis sebagai TransJakarta) adalah sistem transportasi Bus Rapid Transit (BRT) pertama di Asia Tenggara dan Selatan, yang beroperasi sejak tahun 2004 di Jakarta, Indonesia. Sistem ini didesain berdasarkan sistem TransMilenio di Bogota, Kolombia. Transjakarta dirancang sebagai moda transportasi massal pendukung aktivitas ibu kota yang sangat padat. Transjakarta merupakan sistem BRT dengan jalur lintasan terpanjang di dunia (251,2 km), serta memiliki 287 halte yang tersebar dalam 13 koridor (rute utama), yang pada awalnya beroperasi dari pukul 05.00–22.00 WIB, dan kini beroperasi selama 24 jam di seluruh koridornya.

Manfaat Transjakarta :

1. **Tarif Terjangkau**: Transjakarta adalah transportasi umum dengan **tarif paling murah** di Jakarta. Per Oktober 2022, tarifnya adalah:
   - Jam operasional 05.00 – 07.00 WIB: Rp2.000
   - Jam 07.00 – 24.00 WIB: Rp3.500
   - Jam 24.00 – 05.00 WIB: Rp3.500. Dengan tarif ini, warga bisa berpergian ke mana pun di Jakarta tanpa membayar lebih.
2. **Halte Tersedia di Banyak Titik**: Transjakarta memiliki **260 halte** yang tersebar di **13 koridor** di Jakarta. Jadwal kedatangan Transjakarta juga relatif singkat, sekitar lima hingga 12 menit saja.
3. **Fasilitas Memadai**: Bus Transjakarta dilengkapi dengan fasilitas keamanan dan keselamatan, seperti CCTV dan petugas keamanan. Selain itu, fasilitas penumpangnya juga nyaman, termasuk AC, kursi empuk, dan jendela besar.
4. **Meningkatkan Kualitas Hidup**: Dengan menggunakan Transjakarta, masyarakat Jakarta dapat memperoleh akses ke berbagai tempat dengan lebih mudah, sehingga meningkatkan kualitas hidup mereka.
5. **Memperkuat Transportasi Massal di Jakarta**: Dengan menggunakan Transjakarta, masyarakat Jakarta dapat mendukung dan memperkuat sistem transportasi massal yang ada di kota, membantu pemerintah membangun sistem yang lebih baik dan efisien di masa depan².

### **Identifikasi outlier dalam dataset**

```python
numerik = transJ.select_dtypes(include='number')
```

```python
for kolom in numerik.columns:
    q1 = transJ[kolom].describe().loc['25%']
    q3 = transJ[kolom].describe().loc['75%']
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr 

    jumlah_outliers = transJ[(transJ[kolom] < lower_bound) | (transJ[kolom] > upper_bound)].shape[0]
    print(f"data {kolom} memiliki outliers {jumlah_outliers}")
```

```python
fig = plt.figure(figsize=(20,20)) # horizontal, vertikal

count = 0
for kolom in numerik.columns:
    count += 1
    plt.subplot(8,2,count)
    sns.boxplot(numerik, x = kolom)
    fig.tight_layout() # biar tidak tertumpuk
    plt.title(f'{kolom}')
plt.show()
```

#### **Fixing Missing Value**

Mengisi kolom `corridorID` dengan .filna(method='ffill') yang berfungsi untuk mengisi kan missing value berdasarkan data yang berkesinambungan dengan kolom `corridorName` diatas data missing value itu sendiri. Begitu juga sebaliknya dengan kolom `corridorName`.

```python
# Hubungan berkelanjutan antara kolom `corridorName` dan `corridorID` dengan `corridorID` sebagai poros missing value nya
transJ_clean.loc[(transJ_clean['corridorName'] == 'Kampung Rambutan - Blok M') | (transJ_clean['corridorID'].isna()) & (transJ_clean['corridorID']=='7B'), ['corridorName', 'corridorID']]
```

```python
# Hubungan berkelanjutan antara kolom `corridorName` dan `corridorID` dengan `corridorName` sebagai poros pencarian missing value
transJ_clean.loc[(transJ_clean['corridorID'] == '2') | (transJ_clean['corridorName'].isna()) & ((transJ_clean['corridorName']=='Pulo Gadung - Monas')), ['corridorName', 'corridorID']]
```

Dari 2 tabel diatas, bisa disimpulkan semua 'corridorName' dan 'corridorID' memiliki keterkaitan, jadi missing value bisa di isi dengan .fillna(method='ffill') yang fungsinya mengisi missing value dari pola data yang sudah ada dengan groupby corridorID sebagai acuan.

```python
# Penggunaan groupby untuk pola data dari kolom `corridorID` dan `corridorName`, lalu menggunakan transform untuk pengaplikasian lambda .fillna metode ffill dan bfill kepada groupby tadi berdasarkan `corridorName`
transJ_clean['corridorName'] = transJ_clean.groupby('corridorID')['corridorName'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

#### **2. Halte TransJakarta dengan banyaknya penumpang wanita naik dan turun**

   - Pengaturan lalu lintas yang efisien dan penegakan peraturan lalu lintas yang ketat dapat mengurangi kemacetan  dan memastikan kelancaran transportasi umum.

   - Perkirakan biaya pengadaan dan operasional armada tambahan.
   - Bandingkan biaya dengan potensi pendapatan tambahan dari peningkatan jumlah penumpang.
7. **Penyewaan Media Iklan di Halte**:

   - Buka peluang bagi pihak ketiga untuk berinvestasi dalam pengembangan dan pengelolaan fasilitas Transjakarta.


Semua solusi di atas perlu dikoordinasikan dengan pemerintah, operator transportasi, dan masyarakat agar dapat memberikan dampak positif yang signifikan.

Berikut adalah estimasi pendapatan berdasarkan solusi yang telah diusulkan:
1. Penambahan Armada pada Rute Padat:
    * Perhitungan:
        * Jumlah penumpang tambahan per hari = (Jumlah penumpang rata-rata per bus) × (Jumlah armada tambahan)
        * Pendapatan tambahan per hari = (Jumlah penumpang tambahan per hari) × (Tarif tiket per penumpang)
        * Pendapatan tambahan per bulan = (Pendapatan tambahan per hari) × (Jumlah hari operasional per bulan)
    * Contoh:
        * Jika rata-rata penumpang per bus adalah 50 orang, dan kita menambah 10 bus, maka:
            * Jumlah penumpang tambahan per hari = 50 × 10 = 500 orang
            * Pendapatan tambahan per hari = 500 × Rp2.000 = Rp1.000.000
            * Pendapatan tambahan per bulan (30 hari) = Rp1.000.000 × 30 = Rp30.000.000

      Disamping menambah income, dengan penambahan 10 unit bus pada halte terpadat, durasi tunggu kedatangan bus di halte yang sebelumnya adalah 30 menit akan menjadi 15 menit. Ini akan mengurangi antrian penumpang dan mempercepat layanan transportasi Transjakarta. 

2. Penyewaan Media Iklan di Halte:
    * Perhitungan:
        * Jumlah unit iklan = 20 unit
        * Harga iklan per unit = Rp5.000.000
        * Durasi kontrak = 6 bulan
        * Pendapatan dari penyewaan iklan = 20 × Rp5.000.000 × 6 = Rp600.000.000

3. Pengelolaan Pendapatan Non-Tiket (Non-Farebox):
    * Perhitungan:
        * Pendapatan dari naming rights = Rp1.000.000 × 50 = Rp50.000.000
        * Pendapatan dari kegiatan komersial = Rp10.000.000
        * Total pendapatan non-tiket = Rp50.000.000 + Rp10.000.000 = Rp60.000.000

4. Kerjasama dengan Swasta dan Pihak Ketiga:
    * Perhitungan:
        * Investasi dari swasta = Rp5.000.000.000
        * Bagi hasil dari pendapatan = 20% dari pendapatan tambahan per bulan = 0.2 × Rp30.000.000 = Rp6.000.000
        * Pendapatan dari kerjasama dengan swasta = Rp6.000.000 × 12 = Rp72.000.000
   
Jadi, total pendapatan tambahan per bulan dari semua solusi adalah: Rp30.000.000 + Rp600.000.000 + Rp60.000.000 + Rp72.000.000 = Rp762.000.000


Rekomendasi solusi untuk mengatasi antrian panjang saat penumpang hendak tap out di Transjakarta sehingga penumpang tidak lagi melewatkan tap out demi dataset yang lebih akurat :

1. Penyediaan Petugas Tambahan di Halte:

    * Petugas dapat membantu penumpang yang mengalami kesulitan saat tap out.
2. Sistem Tap Out Otomatis:
    * Kembangkan sistem otomatis yang mendeteksi penumpang yang turun dari bus.
    * Penumpang tidak perlu lagi menempelkan kartu secara manual.
3. Sosialisasi Lebih Intensif:
    * Edukasi penumpang mengenai pentingnya tap out dan cara penggunaannya.
    * Sosialisasi melalui pengumuman di dalam bus dan di halte.
4. Penggunaan Teknologi NFC (Near Field Communication):
    * Gunakan teknologi NFC pada kartu elektronik agar proses tap out lebih cepat dan efisien.
